**NLP- Example ffor Single Sentence**

**1. Tokenization**

Tokenization is the process of breaking text into smaller units, such as sentences or words.

**Sentence Tokenization**

This splits a paragraph into individual sentences.

Example:
Input: "John works at Google. He lives in New York and loves programming in Python."

Output: ["John works at Google.", "He lives in New York and loves programming in Python."]

**Word Tokenization**

This splits sentences into individual words.

Example:

Input: "John works at Google."
Output: ["John", "works", "at", "Google", "."]
ðŸ”¹ Why is it important?
Helps in text processing by breaking down data into meaningful elements.

**2. Stopword Removal**

Stopwords are common words that do not add much meaning to a sentence, such as "is," "the," "at," "in," "and," etc.

Removing stopwords helps in reducing noise and improving the efficiency of NLP models.

Example:
Input: "He is working at Google in New York."
Without Stopwords: ["He", "working", "Google", "New York"]

ðŸ”¹ Why is it important?
Reduces data size and processing time.
Focuses on meaningful words.

**3. Lemmatization**

Converts words to their root or dictionary form (lemma).

Unlike stemming, which simply chops off word endings (e.g., "running" â†’ "runn"), lemmatization ensures that the word retains its meaning.

Example:
Word	Lemmatized Form
Running	Run
Better	Good
Studies	Study
ðŸ”¹ Why is it important?

Standardizes words to improve NLP model accuracy.
Avoids redundancy in word forms.

**4. Part-of-Speech (POS) Tagging**

Assigns grammatical categories (like nouns, verbs, adjectives) to words.
Helps in understanding sentence structure.
Example:
Input: "John works at Google."
Output:
[("John", "NNP"), ("works", "VBZ"), ("at", "IN"), ("Google", "NNP")]

NNP (Proper Noun) â†’ "John", "Google"
VBZ (Verb, 3rd person singular present) â†’ "works"
IN (Preposition) â†’ "at"
ðŸ”¹ Why is it important?

Useful in text analysis, information retrieval, and machine translation.
Helps with Named Entity Recognition (NER).

**5. Named Entity Recognition (NER)**

Identifies and categorizes names, locations, organizations, dates, and more.

Example:
Input: "John works at Google in New York."
Output:
[("John", PERSON), ("Google", ORG), ("New York", GPE)]

PERSON â†’ "John"

ORG (Organization) â†’ "Google"

GPE (Geopolitical Entity, like a city or country) â†’ "New York"

ðŸ”¹ Why is it important?
Used in chatbots, search engines, and data extraction.
Helps machines understand real-world entities.

**2. Tokenizers**
NLTK provides tools for breaking text into words and sentences.


**punkt â€“** A pre-trained tokenizer model for sentence splitting.

**word_tokenize â€“** Tokenizes a sentence into words.

**sent_tokenize â€“** Splits a paragraph into sentences.

**3. Stemmers & Lemmatizers**
Used for reducing words to their base forms.

**PorterStemmer â€“** Uses stemming to reduce words to their root form.

**LancasterStemmer â€“** A more aggressive stemming algorithm.

**WordNetLemmatizer**â€“ Uses WordNet to find the base (lemma) of words.

**averaged_perceptron_tagger in NLTK**

The averaged_perceptron_tagger is a pre-trained Part-of-Speech (POS) tagger in NLTK that uses the Averaged Perceptron algorithm to assign grammatical labels (POS tags) to words in a sentence.

**maxent_ne_chunker - Named Entity Chunker**

Uses the Maximum Entropy model to identify named entities in a text.
It is trained on the CoNLL 2000 corpus, a well-known dataset for chunking.
Identifies entities like PERSON, ORGANIZATION, LOCATION, GPE (Geo-Political Entity), DATE, TIME, MONEY, etc.


**maxent_ne_chunker_tab is a tab-separated version of the maxent_ne_chunker model.**
It is internally used by NLTK for:

Storing pre-trained Named Entity Recognition (NER) data.
Retrieving entity-related information in a structured format.

** Key Points:**
It is not directly used by users.
It functions as a supporting resource for maxent_ne_chunker.
It helps in efficient chunking by storing entity rules in a tab-separated format.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, ne_chunk

# Download necessary NLTK resources
nltk.download('punkt')  # Tokenization
nltk.download('stopwords')  # Stopwords for filtering
nltk.download('wordnet')  # Lemmatization
nltk.download('averaged_perceptron_tagger')  # POS tagging
nltk.download('maxent_ne_chunker')  # Named Entity Recognition (NER)
nltk.download('words')  # Dictionary for NER
nltk.download('punkt_tab') # Download the punkt_tab resource
nltk.download('averaged_perceptron_tagger_eng') #download the english version of the tagger
nltk.download('maxent_ne_chunker_tab') # Download the correct tab resource for ne_chunker

# Sample text for NLP processing
text = "The cat meows softly, asking for a tasty sweet treat."

# Step 1: Sentence Tokenization - Splitting text into sentences
sentences = sent_tokenize(text)
print("Sentence Tokenization:", sentences)

# Step 2: Word Tokenization - Splitting sentences into words
words = word_tokenize(text)
print("\nWord Tokenization:", words)

# Step 3: Removing Stopwords - Filtering out common words like 'the', 'is', etc.
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("\nAfter Stopword Removal:", filtered_words)

# Step 4: Lemmatization - Reducing words to their base form
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nAfter Lemmatization:", lemmatized_words)

# Step 5: Part-of-Speech Tagging - Assigning grammatical categories to words
pos_tags = pos_tag(words)
print("\nPOS Tagging:", pos_tags)

# Step 6: Named Entity Recognition (NER) - Identifying names, places, and organizations
ner_tree = ne_chunk(pos_tags)
print("\nNamed Entity Recognition:")
ner_tree.pretty_print()
```

```python
import string

# Get English stopwords
stop_words = set(stopwords.words('english'))

# Get the set of all punctuation marks
punctuation = set(string.punctuation)

# Remove stop words and punctuation from the tokens
data['cleaned_title'] = data['lowercase_title'].apply(lambda x: [word for word in x if word not in stop_words and word not in punctuation])

# Display the cleaned tokens (without stop words and punctuation)
data[['title', 'cleaned_title']].head()
```

```python


import re

# Function to remove special characters from tokens
def remove_special_characters(text):
    # Use regex to keep only alphanumeric characters (letters and numbers)
    cleaned_text = [re.sub(r'\W+', '', word) for word in text if re.sub(r'\W+', '', word) != '']
    return cleaned_text

# Remove stop words, punctuation, and special characters from tokens
data['cleaned_title'] = data['lowercase_title'].apply(lambda x: [word for word in x if word not in stop_words and word not in punctuation])

# Apply the function to remove special characters
data['cleaned_title'] = data['cleaned_title'].apply(remove_special_characters)

# Drop duplicate reviews (if any)
data.drop_duplicates(subset='title', keep='first', inplace=True)

# Display the cleaned tokens (without special characters and duplicates)
data[['title', 'cleaned_title']].head()
```

```python
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to the tokens
data['lemmatized_title'] = data['cleaned_title'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Display the lemmatized tokens
data[['title', 'lemmatized_title']].head()
```

```python
# Download VADER for sentiment analysis
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Apply sentiment analysis to lemmatized reviews
data['sentiment_scores'] = data['tokenized_title'].apply(lambda review: sid.polarity_scores(' '.join(review)))

# Extract compound sentiment score (overall sentiment)
data['compound_score'] = data['sentiment_scores'].apply(lambda score_dict: score_dict['compound'])

# Categorize reviews based on compound score
data['sentiment'] = data['compound_score'].apply(lambda x: 'positive' if x >= 0.05 else ('negative' if x <= -0.05 else 'neutral'))

# Display reviews with their sentiment
data[['title', 'tokenized_title', 'compound_score', 'sentiment']].head()
```

# **Document Silimlarity **

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample documents
documents = [
    "The sun is shining bright in the morning today.",
    "I like ice cream.",
    "I have a dog named Princess.",
    "My dog is black and white.",
]

# Step 1: Convert the documents to TF-IDF vectors
vectorizer = TfidfVectorizer()

# Transform the documents into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(documents)
print(tfidf_matrix)
# Step 2: Compute the Cosine Similarity between the documents
cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Step 3: Display the similarity matrix
print("Cosine Similarity Matrix:")
print(cosine_sim_matrix)
```
